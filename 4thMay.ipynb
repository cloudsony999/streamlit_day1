```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
empdata={ "ename":['rahul','shreya','koyel','dwaipayan'],"salary":[20000,25000,40000,30000]}
```

```python
empdata
```

```python
df=pd.DataFrame(empdata)
```

```python
df
```

```python
x=df['ename']
y=df['salary']
```

```python
x
```

```python
y
```

```python
plt.xlabel('Employee Name')
plt.ylabel('Employee Salary')
plt.title('RAHUL AND JAVA')
plt.bar(x,y,label='Employee Data',color='red')
plt.legend()
plt.show()
```

```python
age=[6,8,9,2,10,2,8,9,11,12,60]
price=[97,87,82,120,77,122,99,97,74,77,83]
plt.xlabel('House AGE in years')
plt.ylabel('Price in Lakhs')
plt.title('House Price')
plt.scatter(age,price,color='red')
plt.show()
```
